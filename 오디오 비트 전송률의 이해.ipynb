# 음악, 음성, 사운드를 오디오 데이터로 변환

음악이나 음성사운드를 데이터로 변환할 때 생기는것이 데이터 품질에 관한 이슈이다.디지털 오디오는 디지털 정보이다. 이 정보는 밀도가 높거나 낮고, 품질이 좋거나 나쁠 수 있다.  

비트 전송률은 오디오로 전송되는 데이터의 양을 의미합니다. 비트 전송률이 높을수록 오디오 품질이 좋아집니다. 프로듀서이자 엔지니어인 Gus Berry는 "비트 전송률이 오디오 피델리티(fidelity)를 결정합니다. 

사운드가 아무리 좋더라도 비트 전송률이 낮다면 사운드 품질은 떨어지게 됩니다"라고 설명합니다. 오디오를 녹음하고 제작 및 배포하려면 비트 전송률에 대한 이해가 필요하며, 비트 전송률을 제대로 이해하려면 먼저 오디오 파일을 구성하는 요소와 다양한 오디오 파일 유형을 알아야 합니다.

# 비압축, 무손실 및 압축 오디오 파일

프로듀서 Peter Rodocker는 "사운드는 파형으로 구성되며, 오디오 파일은 그 파형을 나타냅니다. 개별 샘플을 통해 파형이 오디오 파일에 인코딩되는 방식에는 특정 순간의 파형 모양과 ‘0’에서 멀어지는 정도가 포함됩니다"라고 말합니다. 

‘0’인 경우 무음이 되며, 오디오 파일은 무음에서 사운드의 거리를 측정합니다. "이는 기본적으로 오디오 파형의 스냅샷"이라고 Rodocker는 말합니다. 스냅샷의 유형은 매우 다양합니다. 이미지마다 품질과 선명도가 다른 것처럼 오디오 파일 유형마다 크기와 정보량, 역할이 모두 다릅니다. 

예외 사항도 있지만 비압축 파일에는 가장 많은 정보가 포함되므로 비트 전송률이 가장 높습니다. 압축 손실 파일에는 일반적으로 가장 적은 정보가 포함되므로 비트 전송률이 낮습니다.

    - 비압축 파일: 오디오 파일 용량이 매우 크며, 오디오 장비가 감지할 수 있는 모든 정보를 포함합니다. 비압축 파일 포맷으로는 WAV, AIFF, PCM이 있습니다.

    - 압축 무손실 파일: 압축된 파일이지만 정보가 손실되지 않습니다. 포맷으로는 FLAC, WMA, ALAC가 있습니다. 파일 크기는 압축 파일보다 크고 비압축 파일보다 작습니다.

    - 압축 파일 또는 손실 파일: 일반적으로 가장 작은 파일 유형인 압축 파일은 중요하지 않은 정보를 제거합니다. 흔히 사용되는 손실 오디오 파일 포맷으로는 MP3와 AAC가 있습니다. Rodocker는 "Apple Music과 Spotify 등에서 스트리밍될 때 흔히 사용됩니다"고 말합니다.

# 샘플 속도와 비트 심도
 쌤플 속도는 오디오 샘플이 캡처되는 초당 횟수 즉, 녹음 장비가 사운드를 데이터로 변환하는 초당 인스턴스 수입니다. 대부분 디지털 오디오의 샘플링 속도는 44.1kHz이며, 오디오 CD의 샘플링 속도 또한 이와 같습니다. 즉, 오디오는 녹음하는 동안 초당 44,100회 샘플링됩니다. 오디오가 재생되면 하드웨어가 사운드를 초당 44,100회 재구성합니다. 개별 샘플마다 다른 정보량을 가집니다. 비트 심도는 각 샘플의 비트 수, 즉 44,100개의 각 오디오 조각이 가진 정보량입니다. 샘플 속도와 비트 심도가 높으면 오디오 파일의 정보량이 많아지고 파일 크기도 커집니다. 해상도가 높은 사진처럼 샘플 속도와 비트 심도가 높은 오디오 파일은 세부 정보가 많습니다. 일반적으로 세부 정보를 많이 얻으려면 비트 전송률을 높여야 합니다.

# 최적의 비트 전송률 찾기

새로운 오디오 프로젝트를 시작할 때 높은 샘플 속도와 비트 심도로, 가능한 최고의 품질로 녹음하는 것이 좋습니다. 오디오를 제작하려는 경우 청취자가 오디오를 어떤 방식으로 들을지 생각해봐야 합니다.

 

오디오 유형과 관계없이 최신 기술에도 사용할 수 있도록 최고 품질의 파일을 유지해야 합니다. 향후 오디오 포맷은 높은 비트 전송률을 필요로 할 것이며, 미래형 연결 시스템 또는 하드웨어는 하이 피델리티 오디오를 전달할 것입니다. 상업용 오디오 하드웨어는 앞으로 10년 안에 160Kbps 이상의 비트 전송률을 송출할 것으로 예상되므로, 기존의 비압축 파일을 보관해 두시기 바랍니다. 비압축 파일은 새로운 파일 포맷으로 변환할 수 있기 때문입니다.

 

사용할 파일 유형을 알아두는 것은 오디오 제작의 일부에 불과합니다. 그래픽 이퀄라이저를 사용하는 방법과 음악을 믹싱하는 방법을 학습하면 비트 전송률에 상관없이 오디오를 크게 향상시킬 수 있습니다.
